```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
file_path = 'HRI Project.csv' 
data = pd.read_csv(file_path, header=None) 
```

```python
column_indices = {
    'Q0': 0, 
    'Q1': 1, 
    'Q2': 2, 
    'Q3': 3, 
    'Q4': 4, 
    'Q5': 5,
    'Q6': 6,
    'Q7': 7,
    'Q8': 8,
    'Q9': 9,
    'Q10': 10,
    'Q11': 11,
    'Q12': 12,
    'Q13': 13,
    'Q14': 14,
    'Q15':15,
    'Q16':16,
    'Q17':17,
    'Q18':18,
    'Q19':19,
    'Q21':21,
    'Q22': 22, 
    'Q23': 23, 
    'Q24':24, 
    'Q25': 25,
    'Q26': 26,
    'Q27': 27,
    'Q28': 28,
    'Q29': 29,
    'Q30': 30,
    'Q31': 31, 
    'Q32': 32, 
    'Q33': 33, 
    'Q34': 34, 
    'Q35': 35,
    'Q36': 36,
    'Q37': 37,
    
    
}
```

```python
def process_column(data, column_index):
    # Getting the description from the first row
    description = data.iloc[1, column_index]
    # Getting the next 11 rows as an array, converting to float, replacing non-numeric values with 0
    data_array = pd.to_numeric(data.iloc[1:12, column_index], errors='coerce').fillna(0).to_numpy()
    return description, data_array
```

```python
column_data = {}
for col, index in column_indices.items():
    description, data_array = process_column(data, index)
    column_data[col] = {"Description": description, "Data Array": data_array} 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

# Processing data for Q0 and Q3
Q0_description, Q0_data_array = process_column(data, column_indices['Q0'])
Q3_description, Q3_data_array = process_column(data, column_indices['Q3'])

# Cleaning and converting data for Q0
q0_data_clean = pd.to_numeric(Q0_data_array[~np.isnan(Q0_data_array)], errors='coerce')

# Cleaning and converting data for Q3
q3_data_clean = pd.to_numeric(Q3_data_array[~np.isnan(Q3_data_array)], errors='coerce')

# Perform the t-tests for Q0 and Q3
t_stat_03, p_value_03 = stats.ttest_ind(q0_data_clean, q3_data_clean, equal_var=False)

# Print the t-test results
print(f"T-statistic for Q0 vs Q3: {t_stat_03}, P-value: {p_value_03}")

# Plotting the data for Q0 and Q3
plt.figure(figsize=(10, 6))
plt.plot(q0_data_clean, marker='o', label='Q0 Data', color='blue')
plt.plot(q3_data_clean, marker='x', label='Q3 Data', color='orange')
plt.xlabel('Index')
plt.ylabel('Comparison of Interface Robot and Dialogue Robot')
plt.title('I trust my creative abilities')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Extract the relevant columns 
q0_data = Q0_data_array
q3_data = Q3_data_array


q0_data_clean = q0_data[~np.isnan(q0_data)]
q3_data_clean = q3_data[~np.isnan(q3_data)]


# Convert to numeric, replacing non-numeric values with NaN
q0_data = pd.to_numeric(q0_data, errors='coerce')
q3_data = pd.to_numeric(q3_data, errors='coerce')


# Filter out NaN values
q0_data_clean = q0_data[~np.isnan(q0_data)]
q3_data_clean = q3_data[~np.isnan(q3_data)]


# Perform the t-tests
t_stat_03, p_value_03 = stats.ttest_ind(q0_data_clean, q3_data_clean, equal_var=False)


# Print the results
print(f"T-statistic for Q0 vs Q1: {t_stat_03}, P-value: {p_value_03}")


if p_value_03 < 0.05:
    print("There is a significant difference between the two groups.")
else:
    print("There is no significant difference between the two groups.")
```
